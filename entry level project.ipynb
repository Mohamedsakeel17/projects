```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

print(iris.head())
```

```python
print(iris.info())

print(iris.describe())
```

```python
plt.figure(figsize=(10, 6))

iris.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.suptitle('Distribution of Features', fontsize=16)
plt.show()
```

```python
sns.pairplot(iris, hue='species', height=2.5)
plt.suptitle('Pairwise Relationships between Features', y=1.02, fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris)
plt.title('Box Plot of Features', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='sepal_length', data=iris, inner='quartile')
plt.title('Violin Plot of Sepal Length across Species', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Features', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris, s=100)
plt.title('Sepal Length vs. Petal Length', fontsize=16)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
